### Two option Subplot 

```python
# subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(15,10))
ax1.plot(np.random.rand(10))
ax2.scatter(np.random.rand(10),np.random.rand(10))
ax3.bar(nut_butter_price.keys(),nut_butter_price.values(),color=["Red","Green","Brown"]);
ax3.set(title="Nut Butter Price",
      xlabel="Butter Name",
      ylabel="Butter Price in Dollar");
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))
ax[0,0].plot(np.random.rand(10));
ax[0,1].scatter(np.random.rand(10),np.random.rand(10))
ax[1,0].barh(list(nut_butter_price.keys()),list(nut_butter_price.values()))
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (Pyplot vs OO)

* When plotting something quickly it is ok to use pyplot
* When ploting something something advance use OO

```python
# Lets fillter those patients who's age are greater than 50
over_50= heart_disease[heart_disease["age"]>50]
heart_disease.head()
# len(over_50)
```

```python
over_50.plot.scatter(x="age",
                    y="chol",
                    c="target");
```

We used pyplot method to Quick Scatter plot but here is no x label and no x axis no marking

Lets use Object Oriented Method`

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot.scatter(x="age",
            y="chol",
            c="target",
            ax=ax)
ax.set_xlim([45,100]);
```

```python
# setting up the plot using OO method
fig,ax=plt.subplots(figsize=(10,7))

# passing values in the axes and setting up color
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

# customizing our graph
ax.set(title="Heart disease and cholestrol level",
       xlabel="Age",
       ylabel="Cholestrol");

# setting up legend
ax.legend(*scatter.legend_elements(),title="Target");

# adding a horizental line

ax.axhline(y=over_50["chol"].mean(),
           linestyle="-.");
```

```python
over_50.head()
```

```python
# subplots of chol, age and thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)

# add data to ax0

scatter= ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customise ax0

ax0.set(title="Age and Cholestrol Level",
        ylabel="Cholestrol")

ax0.legend(*scatter.legend_elements(), title="Target");

# mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

# add data to ax1

scatter1=ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# customise ax1

ax1.set(title="Age and Max Heart Rate",
        xlabel="Age",
        ylabel="Thalach");

# Add legend
ax1.legend(*scatter1.legend_elements(), title="Targets");

# Add mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# set the style for the plot
plt.style.use("seaborn-whitegrid") #use plt.style.available to get all the styles name

# subplots of chol, age and thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,  #no of rows
                            ncols=1,  #no of columns
                            figsize=(15,15), #the size of the fig
                            sharex=True) #values of all the x axis are been shared because they contain the same values

# setup the ax0...

# add data to ax0

scatter= ax0.scatter(x=over_50["age"],  #added x value
                     y=over_50["chol"], #added y value
                     c=over_50["target"], #added color value depending on the target
                     cmap="winter") #added colour theme for the plot
# setting up x and y limits
ax0.set_xlim([50,80])
ax0.set_ylim([100,600])

# customise ax0

ax0.set(title="Age and Cholestrol Level", #setting up title
        ylabel="Cholestrol") #setting up ylable


ax0.legend(*scatter.legend_elements(), title="Target"); #setting up the legend

# mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

# setting up the ax1...

# add data to ax1

scatter1=ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# setting up x and y limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# customise ax1

ax1.set(title="Age and Max Heart Rate",
        xlabel="Age",
        ylabel="Thalach");

# Add legend
ax1.legend(*scatter1.legend_elements(), title="Targets");

# Add mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# setting up the fig name
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

# saving the fig in the storage
fig.savefig("Heart Disease Analysis.png")
```
